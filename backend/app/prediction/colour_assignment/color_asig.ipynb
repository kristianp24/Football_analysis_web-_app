```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
image =  cv2.imread('image_for_clustering/player.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#used to convert the color format of the image from BGR (Blue-Green-Red) to RGB (Red-Green-Blue).
plt.imshow(image)
plt.show()
```

```python
top_half = image[:int(image.shape[0]/2), :]

plt.imshow(top_half)
plt.show()
```

```python
# each row should represent a pixel which has 3 of its colors RGB
array_2d = top_half.reshape((-1,3))

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(array_2d)
print(kmeans.labels_)
print(np.bincount(kmeans.labels_)) # points assigned to each cluster


```

```python
labels = kmeans.labels_
image_clustered =  labels.reshape(top_half.shape[0], top_half.shape[1])

plt.imshow(image_clustered)
plt.axis("off")

plt.show()
```

```python
background_cluster = [image_clustered[0,0], image_clustered[0,-1], image_clustered[-1,0], image_clustered[-1, -1]]
print(background_cluster)
player_cluster = 1 - background_cluster[0]
print(player_cluster)
print(kmeans.cluster_centers_[player_cluster])
```
